```python
df_employees = pd.merge(df_hr_employees,df_sales_orders, how= "left",on=["EMPID"])
```

```python
df_employees["BIRTHDATE"] = pd.to_datetime(df_employees["BIRTHDATE"])

df_employees.head(2)
```

```python
#DATETIME IZRAČUN STAROSTI

import datetime
#y = datetime.datetime.now()
y = datetime.datetime(2020, 2, 15)
```

```python
years = []

for year in df_employees["BIRTHDATE"]:    
    i = y - year
    i = i.days
    z = i/365
    
    years.append(z)

```

```python
years
```

```python
# izpišimo leta zaposlenih iz lista v array
leta_zaposlenih = list(map(int,years))
```

```python
# ustvarjanje novega dataframa katerega bomo vnesli v df employees
df_leta_zaposlenih = pd.DataFrame(leta_zaposlenih,columns=["STAROST ZAPOSLENEGA V LETIH"])
```

```python
df_employees["STAROST ZAPOSLENEGA"] = df_leta_zaposlenih
df_employees = df_employees.drop_duplicates(subset="EMPID")
```
