df1 = združevanje tabele "sales_orders" ter "sales_orderdetails" na = ORDERID

```python
df1 = pd.merge(df_sales_orders,df_sales_orderdetails,  how= "left", on=["ORDERID"])
df1.head()
```

df2 = združevanje tabele "productions_category" ter "productions_products" na = CATEGORYID

```python
df2 = pd.merge(df_production_categories, df_production_products,how="left",on=["CATEGORYID"])
df2.head(3)
```

```python
df2 = df2.drop(columns=["UNITPRICE"])
```

```python
df3 = pd.merge(df1,df2,how="left",on=["PRODUCTID"])
df3.head()
```

```python
df3.info() # hiter pregled vseh stolpcev,zapisa in vrednosti, ki se nahajajo v df3.
```

```python
# sprememba zapisa datuma object na datetime 
df3["ORDERDATE"] = pd.to_datetime(df3["ORDERDATE"])
# ustvarjanje novega dataframa, imenovanega df_2015 
df_2015 = df3.copy()
df_2015["YEAR_ORDER_2015"] = df_2015["ORDERDATE"].dt.year 
```

```python
df_2015 = df_2015[df_2015["YEAR_ORDER_2015"]== 2015]
df_2015.head(1)
```

```python
df_2015["SUM"] = df_2015["UNITPRICE"] * df_2015["QTY"] # ustvarili nov stolpec in vanj vnesli ceno za vsak artikel v letu 2015
df_2015[["CATEGORYID","CATEGORYNAME","UNITPRICE","QTY","SUM"]].head(3) # izpis samo željenih stolpcev
rezultat1 = df_2015.groupby(["CATEGORYNAME"]).sum().sort_values("SUM",ascending = False)[["SUM"]] # izpis kategorij in prodaja za leto 2015
rezultat1
```

```python
#skupna prodaja za leto 2015 
df_2015["SUM"].sum()
```

GRAFIČNI PRIKAZ

```python
plt.figure(figsize = (10,10))
sns.barplot(x = rezultat1.index, y= "SUM", data=rezultat1,palette= "Set3")
```

 povprečna cena v kategoriji za leto 2015

```python

rezultat2 = df_2015.groupby(["CATEGORYNAME"])["UNITPRICE"].mean()
rezultat2
```

```python
# število naročil po kategorijah za leto 2015
rezultat3 = df_2015.groupby(["CATEGORYNAME"]).sum()[["QTY"]].sort_values("QTY",ascending=False)
rezultat3

```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = rezultat3.index, y= "QTY", data=rezultat3,palette= "Set2")
plt.show()
```

```python
df_employees = pd.merge(df_hr_employees,df_sales_orders, how= "left",on=["EMPID"])
```

```python
df_employees["BIRTHDATE"] = pd.to_datetime(df_employees["BIRTHDATE"])

df_employees.head(2)
```

```python
#DATETIME IZRAČUN STAROSTI

import datetime
#y = datetime.datetime.now()
y = datetime.datetime(2020, 2, 15)
```

```python
years = []

for year in df_employees["BIRTHDATE"]:    
    i = y - year
    i = i.days
    z = i/365
    
    years.append(z)

```

```python
years
```

```python
# izpišimo leta zaposlenih iz lista v array
leta_zaposlenih = list(map(int,years))
```

```python
# ustvarjanje novega dataframa katerega bomo vnesli v df employees
df_leta_zaposlenih = pd.DataFrame(leta_zaposlenih,columns=["STAROST ZAPOSLENEGA V LETIH"])
```

```python
df_employees["STAROST ZAPOSLENEGA"] = df_leta_zaposlenih
df_employees = df_employees.drop_duplicates(subset="EMPID")
```

```python
starost_do_25 = df_employees[df_employees["STAROST ZAPOSLENEGA"]<26][["FIRSTNAME","STAROST ZAPOSLENEGA"]]
starost_do_25.set_index("FIRSTNAME")
# starost_do_25.set_index("FIRSTNAME").count()  # za željen seštevek !
```

```python
starost_med25_40 =df_employees[(df_employees["STAROST ZAPOSLENEGA"]>25) & (df_employees["STAROST ZAPOSLENEGA"] < 40)][["FIRSTNAME","STAROST ZAPOSLENEGA"]]
starost_med25_40.set_index("FIRSTNAME")
# starost_med25_40.set_index("FIRSTNAME").count()  za željen seštevek !
```

```python
starost_nad40 = df_employees[(df_employees["STAROST ZAPOSLENEGA"]>40) & (df_employees["STAROST ZAPOSLENEGA"] < 60)][["FIRSTNAME","STAROST ZAPOSLENEGA"]]
starost_nad40.set_index("FIRSTNAME")
# starost_nad40.set_index("FIRSTNAME").count()  za željen seštevek !
```
